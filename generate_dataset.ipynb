```python
import numpy as np
```

```python
def gen_two_gaus(mu_1, mu_2, sigma_1, sigma_2):

    gau_1_x = np.random.normal(loc=0.0, scale=sigma_1, size=(n_samples))
    gau_1_y = np.random.normal(loc=0.0, scale=sigma_1, size=(n_samples))
    gau_1 = gau_1_x+gau_1_y*1j+mu_1

    gau_2_x = np.random.normal(loc=0.0, scale=sigma_2, size=(n_samples))
    gau_2_y = np.random.normal(loc=0.0, scale=sigma_2, size=(n_samples))
    gau_2 = gau_2_x+gau_2_y*1j+mu_2
    
    return gau_1, gau_2
```

```python
def plot_two_gaus(gau_1, gau_2, centers):
    plt.figure(figsize=(8,6), dpi=300)
    plt.ylabel('Imaginary')
    plt.xlabel('Real')

    x = [ele.real for ele in gau_1]
    y = [ele.imag for ele in gau_1]
    plt.scatter(x, y, color="r", s=1)
    
    x = [ele.real for ele in gau_2]
    y = [ele.imag for ele in gau_2]
    plt.scatter(x, y, color="b", s=1)
    
    plt.scatter(centers[0].real, centers[0].imag, color="black")
    plt.scatter(centers[1].real, centers[1].imag, color="black")
    
    x = np.asarray(range(1000))/50 - 10
    y = -x*3/4
    plt.scatter(x, y, color="black", s=1)
    
    plt.xlim((-12, 12))
    plt.ylim((-12, 12))
    plt.legend(["Gaussian_1", "Gaussian_2"])
    plt.show()
```

```python
mu_1, mu_2 = 4+3j, -4-3j

sigma_1, sigma_2 = 2, 2

n_samples = 2000

gau_1, gau_2 = gen_two_gaus(mu_1, mu_2, sigma_1, sigma_2)
plot_two_gaus(gau_1, gau_2, [mu_1, mu_2])
```
